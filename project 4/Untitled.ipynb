```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: CSV file load karo (use your own path or name)
data = pd.read_csv("covid_data.csv")  # Example: ['Date', 'Confirmed', 'Recovered', 'Deaths']
```

```python
# Step 3: Date format convert karo
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Step 4: Data check karo
print(data.head())
```

```python
# Step 5: Line Plot - Daily Confirmed Cases
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Confirmed'], label='Confirmed Cases', color='blue')
plt.title('Daily Confirmed COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 6: Area Chart - Confirmed, Recovered, Deaths
plt.figure(figsize=(10, 5))
plt.stackplot(data['Date'], data['Confirmed'], data['Recovered'], data['Deaths'],
              labels=['Confirmed', 'Recovered', 'Deaths'],
              colors=['#1f77b4', '#2ca02c', '#d62728'])
plt.title('COVID-19 Cumulative Stats')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend(loc='upper left')
plt.show()
```

```python
# Step 7: Heatmap - Correlation between features
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Confirmed', 'Recovered', 'Deaths']].corr(), annot=True, cmap='Reds')
plt.title('Correlation Heatmap')
plt.show()
```
