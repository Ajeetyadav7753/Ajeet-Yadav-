```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load Data
df = pd.read_csv("student_data.csv")
```

```python
print(df.describe())
```

```python
# Step 3: Averages
avg_marks = df["Marks"].mean()
avg_attendance = df["Attendance"].mean()
avg_logins = df["Logins"].mean()
```

```python
print(f"Average Marks: {avg_marks}")
print(f"Average Attendance: {avg_attendance}")
print(f"Average Logins: {avg_logins}")
```

```python
# Step 4: Correlation Matrix
correlation = df[[ "Attendance","Marks", "Logins"]].corr()
print("\nCorrelation:\n", correlation)
```

```python
# Heatmap of Correlation
plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 5: Top vs Struggling Students
top_students = df[df["Marks"] >= 80]
struggling_students = df[df["Marks"] < 60]
```

```python
# Bar Chart - Top Students
plt.figure(figsize=(6,3))
sns.barplot(x="Name", y="Marks", data=top_students, palette="Greens")
plt.title("Top Students (Marks >= 80)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Chart - Struggling Students
plt.figure(figsize=(6,3))
sns.barplot(x="Name", y="Marks", data=struggling_students, palette="Reds")
plt.title("Struggling Students (Marks < 60)")
plt.xticks(rotation=45)
plt.show()
```
