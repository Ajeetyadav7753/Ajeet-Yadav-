```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("restaurants_data.csv")  
df.head()
```

```python
print(df.columns)

sns.histplot(df['Aggregate rating'], bins=10, kde=True)
plt.title("Distribution of Aggregate rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df['rating_range'] = pd.cut(df['Aggregate rating'], bins=bins, labels=labels)
most_common = df['rating_range'].value_counts().idxmax()
print("Most common rating range:", most_common)
```

```python
average_votes = df['Votes'].mean()
print("Average number of votes per restaurant:", round(average_votes, 2))
```
